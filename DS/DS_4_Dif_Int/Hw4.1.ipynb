```python
import numpy as np
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
from scipy import integrate
```

```python
speed = [25, 35, 45, 30, 60, 120, 100, 100, 70, 75, 80, 65]
time = np.linspace(0, 11, 12)
time
```

```python
plt.plot(time, speed)
plt.grid()
plt.ylim(0,130)
plt.xlim(0,11)
plt.show()
```

```python
time_int = np.linspace(0, 11, 10000)
f = interp1d(time, speed, kind='cubic')
plt.plot(time, speed, 'o')
plt.plot(time_int, f(time_int))
plt.grid()
plt.ylim(0,130)
plt.xlim(0,11)
plt.show()
```

```python
f = interp1d(time, speed, kind='quadratic')
plt.plot(time, speed, 'o')
plt.plot(time_int, f(time_int))
plt.grid()
plt.ylim(0,130)
plt.xlim(0,11)
plt.show()
```

```python
distance = integrate.quad(f, 0, 11)
distance
```
